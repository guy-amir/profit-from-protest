```python
def process_cei_pdfs(pdf_file_list):
    """
    Processes a list of CEI rating PDFs into a list of dataframes, each tagged with its year.
    
    Args:
        pdf_file_list (list[str]): List of file paths to PDF files.
    
    Returns:
        list[pd.DataFrame]: A list of dataframes with a 'Year' column added.
    """
    dataframes = []
    
    for pdf_path in pdf_file_list:
        try:
            year = utils.extract_year_from_filename(pdf_path)
            df = utils.pdf_to_df(pdf_path)
            df['Year'] = year
            dataframes.append(df)
        except Exception as e:
            print(f"[ERROR] Failed processing {pdf_path}: {e}")
    
    return dataframes
```
